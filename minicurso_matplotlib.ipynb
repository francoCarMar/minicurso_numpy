```python
# Crear un vector y representar un grafic olinea con sus valores
import matplotlib.pyplot as plt
import numpy as np

myarray = np.array([1,2,3])
plt.plot(myarray)
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.show()
```

```python
# Crear 2 vectores y representar los puntos en un grafico de dispersion
import matplotlib.pyplot as plt
import numpy as np

myarray = np.array([1,2,3])
myarray2 = np.array([2,3,4])

plt.scatter(myarray, myarray2)
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.show()
```
